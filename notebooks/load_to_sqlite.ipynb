# Load_to_sqlite.ipynb

```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
clean = pd.read_csv('../data/superstore_cleaned.csv') 
engine = create_engine('sqlite:///../data/superstore.db', echo=False)
clean.to_sql('orders', engine, if_exists='replace', index=False)
```

```python
query = """
SELECT category, SUM(sales) AS total_sales, SUM(profit) AS total_profit
FROM orders
GROUP BY category
ORDER BY total_profit DESC
LIMIT 10;
"""
```

```python
df_top_categories = pd.read_sql(query, engine)
```
