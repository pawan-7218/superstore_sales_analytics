```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

# Load cleaned data

```python
df = pd.read_csv('../data/superstore_cleaned.csv')
df['order_month_dt'] = pd.to_datetime(df['order_month'])
```

# Aggregate monthly sales

```python
monthly = df.groupby('order_month_dt').sales.sum().reset_index().sort_values('order_month_dt')
monthly['month_num'] = np.arange(len(monthly))  # integer feature for months
X = monthly[['month_num']]
y = monthly['sales']
```

# Fit linear regression

```python
model = LinearRegression()
model.fit(X, y)
```

# Predict next 3 months

```python
future_months = np.array([[len(monthly)], [len(monthly)+1], [len(monthly)+2]])
preds = model.predict(future_months)
```


# Plot actual vs forecast

```python
plt.figure(figsize=(10,5))
plt.plot(monthly['order_month_dt'], y, label='Actual Sales')
future_dates = pd.date_range(start=monthly['order_month_dt'].max() + pd.offsets.MonthBegin(), periods=3, freq='M')
plt.plot(future_dates, preds, marker='o', linestyle='--', label='Forecast')
plt.title('Monthly Sales Forecast (Linear Regression Baseline)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Create a table of predicted sales
pred_table = pd.DataFrame({
    'Month': future_dates.strftime('%Y-%m'),
    'Predicted_Sales': preds
})

print(pred_table)
```
