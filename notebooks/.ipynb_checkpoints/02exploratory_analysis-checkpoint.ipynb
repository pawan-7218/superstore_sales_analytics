```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mplcursors
%matplotlib inline
%matplotlib widget
```

```python
df = pd.read_csv('../data/superstore_cleaned.csv')
```

# 1. Sales over time

```python
sales_time = df.groupby('order_month').sales.sum().reset_index()

plt.figure(figsize=(12,5))
line, = plt.plot(sales_time['order_month'], sales_time['sales'], marker='o')
plt.xticks(rotation=45)
plt.title('Sales over time')
plt.xlabel('Order month')
plt.ylabel('Sales')
plt.tight_layout()

# Add hover tooltips
cursor = mplcursors.cursor(line, hover=True)
cursor.connect("add", lambda sel: sel.annotation.set_text(
    f"Month: {sales_time['order_month'][sel.index]}\nSales: {sales_time['sales'][sel.index]:,.2f}"
))

plt.show()
```

# 2. Sales vs Profit by Category

```python
cat = df.groupby('category').agg({'sales':'sum','profit':'sum'}).reset_index()
cat_melt = cat.melt(id_vars='category', value_vars=['sales','profit'])

plt.figure(figsize=(8,5))
bars = sns.barplot(data=cat_melt, x='category', y='value', hue='variable')

cursor = mplcursors.cursor(bars, hover=True)
cursor.connect("add", lambda sel: sel.annotation.set_text(
    f"{cat_melt['category'][sel.index]}\n{cat_melt['variable'][sel.index]}: {cat_melt['value'][sel.index]:,.2f}"
))

plt.title('Sales and Profit by Category')
plt.ylabel('Value')
plt.show()
```


# 3. Top 10 Products by Sales

```python
import textwrap
prod = df.groupby('product_name').sales.sum().reset_index().sort_values('sales', ascending=False).head(10)

prod['product_name_wrapped'] = prod['product_name'].apply(lambda x: '\n'.join(textwrap.wrap(x, 25)))

plt.figure(figsize=(16,6))
bars = sns.barplot(data=prod, x='sales', y='product_name_wrapped')

cursor = mplcursors.cursor(bars, hover=True)
cursor.connect("add", lambda sel: sel.annotation.set_text(
    f"{prod['product_name'][sel.index]}: {prod['sales'][sel.index]:,.2f}"
))

plt.title('Top 10 Products by Sales')
plt.show()
```
